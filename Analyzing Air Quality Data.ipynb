```python
# Compare air quality indices between different cities
cities = df['City'].unique()
plt.figure(figsize=(14, 7))
for city in cities:
    city_data = df[df['City'] == city]
    plt.plot(city_data.index, city_data['AQI'], label=city)
plt.legend(loc='upper right',fontsize="7" )
plt.show()
```

```python
# Identify cities with the highest and lowest air quality
avg_aqi=df.groupby("City")["AQI"].mean()
sort_aqi=avg_aqi.sort_values()
max_aqi=sort_aqi.head(1)
min_aqi=sort_aqi.tail(1)
print(max_aqi)
print(min_aqi)
```

```python
# Analyze the data to see if there are any cities with significant seasonal variations.
df["year"]=df["Date"].dt.year
df["month"]=df["Date"].dt.month
monthly_aqi=df.groupby(["City","year","month"])["AQI"].mean().reset_index()
pivot_table=monthly_aqi.pivot_table(index=["year","month"],columns="City",values='AQI')
pivot_table.plot(kind="line",marker="o",linestyle="-")
plt.show()
```

```python
# Monthly Average AQI per City
monthly_aqi = df.groupby(['City', df['Date'].dt.to_period('M')])['AQI'].mean().reset_index()
monthly_aqi.rename(columns={'date': 'month'}, inplace=True)
print(monthly_aqi.to_string())
```

```python
#  Yearly Average AQI per City
yearly_aqi = df.groupby(['City', df['Date'].dt.to_period('Y')])['AQI'].mean().reset_index()
yearly_aqi.rename(columns={'date': 'month'}, inplace=True)
print(yearly_aqi.to_string())
```
